```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [16, 10]
```

```python
def format_date(mdy):
    m, d, y = mdy.split('/')
    if len(m) == 1:
        m = '0' + m
    if len(d) == 1:
        d = '0' + d
    if len(y) == 2:
        y = '20' + y
    return f"{y}-{m}-{d}"

def read(filename):
    df = pd.read_csv(filename)
    df.drop(["Lat", "Long"], "columns", inplace=True)
    df = df.groupby("Country/Region").sum()
    df.rename(columns=format_date, inplace=True)
    df = df.T
    return df
```

```python
confirmed = read("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
deaths = read("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = read("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
df = confirmed - recovered
```

```python
df.loc["2020-02-14":, ["China", "US", "Italy", "Spain", "Germany", "Iran", "France"]].plot(grid=True)
```
